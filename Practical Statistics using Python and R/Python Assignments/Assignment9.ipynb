```python
# read the file
f = open("pubmed_result.txt", "r")


number = 1
for line in f:
    complete = False
    line = f.readline()
    authors = []
    title = ""
    journal= ""
    volume = ""
    issue = ""
    year = ""
    pages = ""

    while not complete:
        if 'FAU - ' in line:
            authors.append(line.replace('FAU - ', '').replace('\n', ''))
        elif 'TI  - ' in line: 
            title = title + line.replace('TI  - ' , '').replace('\n','')
            line = f.readline()
            while '      ' in line:
                title = title + line.replace('      ', '').replace('\n','')
                line = f.readline()
        elif 'JT  - ' in line:
            journal = line.replace('JT  - ' , '').replace('\n','')
        elif 'VI  - ' in line: 
            volume = line.replace('VI  - ', '').replace('\n','')
        elif 'IP  - ' in line: 
            issue = 'no. ' + line.replace('IP  - ', '').replace('\n','')
        elif 'DP  - ' in line: 
            year = line.replace('DP  - ', '').replace('\n','').split(' ')
        elif 'PG  - ' in line: 
            pages = ': ' + line.replace('PG  - ', '').replace('\n','')
        elif '\n' == line or line == '':
            # end of entry
            complete = True
            
            
        line = f.readline()
    # reformat authors
    string_authors = ''
    for author in authors:
        if string_authors == '':
            string_authors = author
        else:
            string_authors = string_authors + '; ' +  author 

    print('[{}] {}. "{}" {} {}, {} ({}){}'.format(number, string_authors, title, journal, volume, issue, year[0], pages))
    number = number + 1
f.close()
```
